```python
df = pd.DataFrame({
    "Customer": ["John", "Sara", np.nan, "Lily", "David", "Emma", "Chris", "Sophia"],
    "Age": [28, np.nan, np.nan, 40, np.nan, 30, 45, np.nan],
    "City": ["New York", "London", np.nan, "Sydney", "Paris", np.nan, "Berlin", "Tokyo"],
    "Purchase_Amount": [250, 400, np.nan, 150, 500, 600, np.nan, 350],
    "Feedback_Score": [5, np.nan, np.nan, 3, 5, np.nan, 2, 4]
})
df=pd.DataFrame(df)
df
```

```python
df.dropna(axis=0,how='all')
```

```python
df.dropna(axis=0,how='any')
```

```python
df.dropna(axis=1,how='all')
```

```python
df.dropna(axis=1,how='any')
```

```python
df['Age'].fillna(20)
```

```python
df['Age'].fillna(df["Age"].median(),inplace=True)
df
```

```python
df['Age'].fillna(method='ffill')
```

```python
df['Age'].fillna(method='bfill')
```

```python
##### Null value treatment
  #### i will check null values
  ### Take percentage of null values
  ### confirm the threshold
  ### if it is beyond threshold drop the columns
  ### if it is less than threshold keep the columns
    #### if it is numercial : mean , medain
    #### if it is categorical : mode
    #### if it is data , time basesd : ffill or bfill
```

```python
  ### Take percentage of null values
null=(df.isna().sum()/df.shape[0])*100
null
```
