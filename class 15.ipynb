```python
s=df.groupby('Order Date')['Sales'].sum()
s.head()
```

```python
plt.plot(s.index,s.values)
plt.xticks(rotation=90)
plt.title("finding the treand of data sales in the order Date")
plt.xlabel("orderdate",fontsize=10)
plt.ylabel("Sales",fontsize=10)
plt.show
```

```python
s=df.groupby('Order Date')['Sales'].sum()
plt.plot(s.index,s.values,color='r',linestyle='--')
plt.xticks(rotation=90)
plt.title("finding the treand of data sales in the order Date")
plt.xlabel("orderdate",fontsize=10)
plt.ylabel("Sales",fontsize=10)
plt.grid(True,linestyle="--")#it is used to style the background lines
plt.show
```

```python
s=df.groupby('Order Date')['Profit'].sum()
plt.plot(s.index,s.values,color='b',linestyle='--')
plt.xticks(rotation=90)
plt.title("finding the treand of Profit in the order Date")
plt.xlabel("orderdate",fontsize=10)
plt.ylabel("Profit",fontsize=10)
plt.grid(True,linestyle="--")#it is used to style the background lines
plt.show
```

```python
s=df.groupby('Region')['Sales'].sum()
plt.bar(s.index,s.values,color='b')
# plt.xticks(rotation=90)
plt.title("region wise sales")
plt.xlabel("region",fontsize=10)
plt.ylabel("Sales",fontsize=10)
plt.grid(True,linestyle="--")#it is used to style the background lines
plt.show
```

```python
s=df.groupby('Region')['Sales'].sum().sort_values()#ascending
plt.bar(s.index,s.values,color='b')
# plt.xticks(rotation=90)
plt.title("region wise sales")
plt.xlabel("region",fontsize=10)
plt.ylabel("Sales",fontsize=10)
plt.grid(True,linestyle="--")#it is used to style the background lines
plt.show
```

```python
s=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)#Descending
plt.bar(s.index,s.values,color=['b','r'])
# plt.xticks(rotation=90)
plt.title("region wise sales")
plt.xlabel("region",fontsize=10)
plt.ylabel("Sales",fontsize=10)
plt.grid(True,linestyle="--")#it is used to style the background lines
plt.show
```

```python
s=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)#Descending
plt.barh(s.index,s.values,color=['b','r'])#barh for horizontal graph 
# plt.xticks(rotation=90)
plt.title("region wise sales")
plt.xlabel("region",fontsize=10)
plt.ylabel("Sales",fontsize=10)
plt.grid(True,linestyle="--")#it is used to style the background lines
plt.show
```

```python
s=df.groupby('Region')['Sales'].sum()
sp=df.groupby('Region')['Profit'].sum()

plt.bar(s.index,s.values,color='r')
plt.bar(sp.index,sp.values,bottom=s.values,color='b',label="Profit")
#these is ccalled stack bar graph
plt.title("region wise sales")
plt.xlabel("region",fontsize=10)
plt.ylabel("Sales",fontsize=10)
plt.grid(True,linestyle="--")#it is used to style the background lines
plt.show
```

```python
plt.scatter(df['Sales'],df['Profit'],marker='^',alpha=0.3,color='r',edgecolors='b')
plt.title("sales and profit")
plt.xlabel("Sales")
plt.ylabel("profit")
```
